```python
# Importing nessesary python libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
#Loading the dataset

df = pd.read_csv("Mens_Shoes.csv")
df.head()
```

```python
 # Checking the dimesion of the dataset
df.shape
```

```python
#Checking the information about the data-set
df.info()
```

```python
# checking number of null values each column contains
df.isnull().sum()
```

```python
# To find the percentage of null values
null_pct = df.isnull().sum()/df.shape[0]*100 
null_pct
```

```python
# columns which contain null values more than 70%
drop_column = null_pct[null_pct>70].keys()
```

```python
#drop columns which cointain 70% of null values
#drop columns that have no use for analysis
df2 = df.drop(columns = drop_column)
drop_column
```

```python
df2.shape
```

```python
#all columns that did not meet the less than 70% null values have been dropped
df2.isnull().sum()
```

```python
# Deleting the rows that contains null values.
df3= df2.dropna() 
```

```python
# Null row values have been removed
df3.isnull().sum()
```

```python
# Renaming columns so they have descriptive names

df4 = df3.rename(columns= {'manufacturerNumber':'manufacturer_Id','name':'Shoe_type','prices.amountMax':'prices',
                          'prices.condition':'Shoe_condition','prices.isSale':'price_status','dateUpdated':'date'})
```

```python
df4
```

```python
#Dropping irrelevent columns
df5 = df4.drop(columns=['prices.merchant','prices.offer','merchants'])
df5.shape
```

```python
df5.dtypes
```

```python
#Changing data types of columns to the correct type

column_datatype = {
    'brand': str,
    'categories': str,
    'manufacturer': str,
    'colors': str,
    'Shoe_type': str,
    'prices': float,
    'Shoe_condition': str,
    'date':'datetime64',
}

df6 = df5.astype(column_datatype)

df6.dtypes

```
